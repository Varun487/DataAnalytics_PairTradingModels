# Create Model for Bajaj
- Test train split

```python
bajaj_df = bajaj_df.reset_index()
bajaj_df.head()
```

```python
# Cropping dataset to be in training range
bajaj_df_in_training_range = bajaj_df[bajaj_df["Date"] < "2018-12-12"]
bajaj_df_in_training_range = bajaj_df_in_training_range[bajaj_df_in_training_range["Date"] > "2018-06-09"]

bajaj_df_in_training_range
```

```python
# Independent variables
features = ['High', 'Low', 'Open', 'Volume']
X = bajaj_df_in_training_range[features]

# Dependent variable (to predict)
y = bajaj_df_in_training_range["Close"]
```

```python
X
```

```python
# Removing last row of X
X = X[:-1]
X
```

```python
y
```

```python
# Removing first row of y
y = y[1:]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=99)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
pair1_regression_model = linear_model.LinearRegression()
pair1_regression_model = pair1_regression_model.fit(X, y)
```

```python
pair1_regression_model.intercept_
```

```python
results = pair1_regression_model.predict(X_test)
```

```python
results
```

```python
print('Intercept: ', pair1_regression_model.intercept_)
print('Coefficients: ', pair1_regression_model.coef_)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, results))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, results))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, results)))
print('R2 Score:', metrics.r2_score(y_test, results))
```
